# **Hand Written Digit prediction- Classification Analysis**

The digits dataset consists of 8* 8 pixel images of digits.The images attribute of the dataset stores 8* 8 arrays of grayscale values for each image.We will use these arrays to visualize the first 4 images.The target attribute of the dataset of the dataset stores the digit each image represents.

# **Import library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Import Data**

```python
from sklearn.datasets import load_digits
```

```python
df=load_digits()
```

```python
_, axes =plt.subplots(nrows=1, ncols=4, figsize=(10,3))
for ax, image, label in zip(axes, df.images, df.target):
    ax.set_axis_off()
    ax.imshow(image, cmap=plt.cm.gray_r, interpolation="nearest")
    ax.set_title("Training:%i" %label)
```

# **Data Preprocessing**

```python
df.images.shape
```

```python
df.images[0]
```

```python
df.images[0].shape
```

```python
len(df.images)
```

```python
n_samples =len(df.images)
data =df.images.reshape((n_samples,-1))
```

```python
data[0]
```

```python
data[0].shape
```

```python
data.shape
```

## **Scalling Image Data**

```python
data.min()
```

```python
data.max()
```

```python
data = data/16
```

```python
data.min()
```

```python
data.max()
```

```python
data[0]
```

## **Train Test Split Data**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(data, df.target, test_size=0.3)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# **Random Forest Model**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

# **Predict Test Data**

```python
y_pred =rf.predict(X_test)
```

```python
y_pred
```

# **Model Accuracy**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
